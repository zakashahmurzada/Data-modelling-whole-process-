```python
correlation_matrix = int_columns.corr()

# Calculate the average correlation with the target variable
average_correlation = correlation_matrix['Attrition'].abs().mean()

# Select features with correlation above the average
selected_features = correlation_matrix[correlation_matrix['Attrition'].abs() > average_correlation].index
print(selected_features)
int_columns = int_columns[selected_features]
```

```python
print(object_columns[1:0])
print(object_columns.BusinessTravel.unique())
object_columns["BusinessTravel"].replace(["Travel_Rarely", "Travel_Frequently", "Non-Travel"], [0, 1, 2], inplace=True)


print(object_columns.Department.unique())
object_columns["Department"].replace(["Sales", "Research & Development", "Human Resources"], [0, 1, 2], inplace=True)

print(object_columns.EducationField.unique()) #drop this
object_columns.drop('EducationField', axis=1, inplace=True)

print(object_columns.Gender.unique())
object_columns["Gender"].replace(["Female", "Male"], [0, 1], inplace=True)

print(object_columns.JobRole.unique()) #drop this
object_columns.drop('JobRole', axis=1, inplace=True)

print(object_columns.MaritalStatus.unique())
object_columns["MaritalStatus"].replace(["Single", "Married", "Divorced"], [0, 1, 2], inplace=True)

print(object_columns.Over18.unique()) #drop this 
object_columns.drop('Over18', axis=1, inplace=True)

print(object_columns.OverTime.unique()) #saxla this
object_columns["OverTime"].replace(["Yes", "No"], [1, 0], inplace=True)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

def get_acc(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = RandomForestClassifier(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    acc_score = accuracy_score(preds_val, val_y)
    return(acc_score)
```

```python
y = general_df['Attrition']
X = general_df.drop('Attrition', axis=1)

from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0, test_size=0.25)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 100, 500, 300]:
    my_acc = get_acc(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(my_acc)
```
